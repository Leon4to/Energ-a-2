```python
df_pais = df_grl.loc[df_grl['pais']=='Afghanistan']
df_pais.cons_energ_percapita.skew()
```

```python
kurtosis_resultados = {'pais': [], 'Kurtosis': []}


for e in df_grl['pais'].unique():  
    datos_pais = df_grl[df_grl['pais'] == e]['cons_energ_percapita']  
    kurtosis = datos_pais.kurtosis()
    kurtosis_resultados['pais'].append(e)
    kurtosis_resultados['Kurtosis'].append(kurtosis)

df_kurto = pd.DataFrame(kurtosis_resultados)

df_kurto
```

```python
kurtosis_resultados = {'continente': [], 'Kurtosis': []}


for e in df_grl['continente'].unique():  
    datos_continente = df_grl[df_grl['continente'] == e]['cons_energ_percapita']  
    kurtosis = datos_continente.kurtosis()
    kurtosis_resultados['continente'].append(e)
    kurtosis_resultados['Kurtosis'].append(kurtosis)

df_kurto_continente = pd.DataFrame(kurtosis_resultados)

df_kurto_continente
```

```python
skewness_resultados = {'pais': [], 'skewness': []}


for e in df_grl['pais'].unique():  
    datos_pais = df_grl[df_grl['pais'] == e]['cons_energ_percapita']  
    skewness = datos_pais.skew()
    skewness_resultados['pais'].append(e)
    skewness_resultados['skewness'].append(skewness)

df_skew = pd.DataFrame(skewness_resultados)

df_skew
```

```python
skewness_resultados = {'continente': [], 'skewness': []}


for e in df_grl['continente'].unique():  
    datos_continente = df_grl[df_grl['continente'] == e]['cons_energ_percapita']  
    skewness = datos_continente.skew()
    skewness_resultados['continente'].append(e)
    skewness_resultados['skewness'].append(skewness)

df_skew_continente = pd.DataFrame(skewness_resultados)

df_skew_continente
```
